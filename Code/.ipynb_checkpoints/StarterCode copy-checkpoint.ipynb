```python
def list_of_ing(x):
    s0 = []
    for ing in x:
        ll = ing.split()
        s = ''
        for word in ll:
            s = s + word + '#'
        #s = '_'.join(ll)
        s0.append(s[:-1])
    return ','.join(s0)

def clean_list(lst):
    out = []
    for item in lst:
        item = re.sub('\d', '', item)
        out.append(item.replace(' ', '_'))
    return ', '.join(out)

def clean_list2(lst):
    #Currently using this version of text cleanup for modeling
    out = []
    for item in lst:
        item = re.sub('\d', '', item)
        item = ''.join(re.findall('[a-zA-Z]', item))
        out.append(item.replace(' ', '_'))
    return ','.join(out)

def clean_list2_1(lst):
    out = []
    for item in lst:
        item = item.lower()
        item = re.sub('\d', '', item)
        item = re.sub('\band\b', ' ', item)
        item = ''.join(re.findall('[a-zA-Z]', item))
        out.append(item.replace(' ', '_'))
    return ','.join(out)

def clean_list3(lst):
    out = []
    for item in lst:
        item = re.sub('\d', '', item)
        item = ''.join(re.findall('[a-z A-Z]', item)).rstrip('_')
        item = item.strip('_')
        out.append(item.replace(' ', '|'))
    return ' , '.join(out)

def clean_list4(lst):
    out = []
    for item in lst:
        item = item.lower()
        item = item.replace("added", "")
        item = item.replace("sauce", "")
        item = item.replace("filets", "")
        item = item.replace("fillets", "")
        item = item.replace("paste", "")
        item = item.replace("added", "")
        item = item.replace("addicted", "")
        item = item.replace("seasoning", "")
        item = item.replace("and", "")
        item = item.replace("boneless", "")
        item = item.replace("dressing", "")
        item = item.replace("skinless", "")
        item = item.replace("slices", "")
        item = item.replace("slices", "")
        item = item.replace("juice", "")
        item = item.replace("mix", "")                                    
        
        item = re.sub('\s+',' ',item)
        item = re.sub('\d', '', item)
        #item = "".join(re.findall("[a-z A-Z]", item)).rstrip('_')
        #item = item.rstrip('_')
        
        out.append(item.replace(" ", "").rstrip(''))                
    return ' , '.join(out)
```

```python
#combine train data into a single string
df_train['text'] = df_train['ingredients'].map(list_of_ing)
```

```python
#combine test data into a single string
df_test['text'] = df_test['ingredients'].map(list_of_ing)
```

```python
#check modified text
df_train.head()
```

```python
#map cuisine labels to integers
tmp_names = list(zip(df_train['cuisine'].value_counts().index, range(1,21)))
dct = {tup[0]: tup[1] for tup in tmp_names}
dct
```

```python
#reverse dictionary for mapping predictions to cuisine names for test data set
reverse_dct = {tup[1]: tup[0] for tup in tmp_names}
reverse_dct
```

```python
#list of cuisine names for classification report
clf_rep_names = df_train['cuisine'].value_counts().index
clf_rep_names
```

```python
# create X,y variables
X = df_train['text']
y = df_train['cuisine'].map(dct)
```

```python
X.head()
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
# Instantiate Count Vectorizer
cvec1 = CountVectorizer()
# instantiate Logistic Regression
lr = LogisticRegression(max_iter=1000)
pipe1 = Pipeline([('cvec', cvec1),
                ('lr', lr)])

# Fit the model
pipe1.fit(X_train, y_train)
print('lr with cvec train score:', pipe1.score(X_train, y_train))
print('lr with cvec test score:', pipe1.score(X_test, y_test))
```

```python
#generate predictions
y_pred = pipe1.predict(X_test)
```

```python
#generate classification report
print(classification_report(y_test, y_pred, target_names=clf_rep_names))
```
